```python
import pandas as pd
```

```python
df = pd.read_csv('small_corpus_with_opinion_sentiment.csv')
```

```python
df.head()
```

```python
def sentiment_score2sentiment_value(sentiment_score):
  if sentiment_score > 0.2:
    return 'positive'
  elif -0.2 <= sentiment_score <= 0.2:
    return 'neutral'
  elif sentiment_score < -0.2:
    return 'negative'
  elif sentiment_score is None:
    return None
```

```python
df['sentiment_value'] = df['sentiment_scores'].apply(lambda x: sentiment_score2sentiment_value(x))
```

```python
df.head(50)
```

```python
# looks like it works well
```

```python
def review_ratings2_rating_classes(review_rating):
  if review_rating == 5:
    return 'positive'
  elif 2 <= review_rating <= 4:
    return 'neutral'
  elif review_rating == 1:
    return 'negative'
```

```python
df['ratings_classes'] = df['ratings'].apply(lambda x: review_ratings2_rating_classes(x))
```

```python
df.head(50)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(df['ratings_classes'], df['sentiment_value'].astype(str))
```

```python
# we can see the accuracy score is 53.25% - not that bad considering there are a 3 classes over here, and we should improve with Nueral-Network
```

```python
from sklearn.metrics import precision_score

precision_score(df['ratings_classes'], df['sentiment_value'].astype(str), average=None)
```

```python
df['ratings_classes'].unique()
```

```python
# unclear why we have 4 categories, maybe the None yields another category (even though it is in the y_pred and not y_true)
```

```python
precision_score(df['ratings_classes'], df['sentiment_value'].astype(str), average=None, labels=['negative', 'neutral', 'positive', 'None'])
```

```python
# precision is: tp/ (tp + fp) = tp / "selected"
# we can see we catch better positive ratings (when fp becomes smaller, when we say it is positive it is more probable to truly be positive) - the last element (0) doesn't mean anything.
```

```python
from sklearn.metrics import recall_score

recall_score(df['ratings_classes'], df['sentiment_value'].astype(str), average=None, labels=['negative', 'neutral', 'positive', 'None'])
```

```python
# recall is: tp / (tp + fn) = tp / "target"
# when fn becomes smaller this means that we are more accurate in detecting the positive objects
# The neutral recall is pretty low, which means we are missing it a lot. However, the negative and positive recall is pretty good (it is also interesting to notice the recall of the positive is higher than its precision - 
# yielding a high f-score)
```

```python
from sklearn.metrics import classification_report

print(classification_report(df['ratings_classes'], df['sentiment_value'].astype(str), labels=['negative', 'neutral', 'positive']))  # , output_dict=True)
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np
from itertools import product
import altair as alt
alt.renderers.enable('notebook')

cm = confusion_matrix(y_true, y_pred, labels=['negative', 'neutral', 'positive'])
cm = [y for i in cm for y in i]
roll = list(product(np.unique(y_true), repeat = 2))
columns = ["actual", "predicted", "confusion_matrix"]
df = pd.DataFrame(columns=columns)
for i in range(len(roll)):
    df = df.append({'actual':roll[i][0], 'predicted':roll[i][1], 'confusion_matrix':cm[i]}, ignore_index=True)
#plot figure
def make_example(selector):
    return alt.Chart(df).mark_rect().encode(
        x="predicted:N",
        y="actual:N",
        color=alt.condition(selector, 'confusion_matrix', alt.value('lightgray'))
    ).properties(
        width=600,
        height=480
    ).add_selection(
        selector
    )
interval_x = alt.selection_interval(encodings=['x'], empty='none')
make_example(interval_x)
# chart.show()
```
