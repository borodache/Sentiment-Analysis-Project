```python
import pandas as pd
```

```python
df = pd.read_csv('small_corpus.csv')
```

```python
df = df.dropna()
df.shape
```

```python
reviews = df['reviews'].dropna().values.tolist()
len(reviews)
```

```python
classifications = classifier(reviews, truncation=True)
```

```python
classifications
```

```python
classifications[-10:]
```

```python
classifications_by_category = []
for cs in classifications:
  if cs['score'] <= 0.75:
    classifications_by_category.append("neutral")
  else:
    classifications_by_category.append(cs['label'].lower())
```

```python
df['classification 75% treshold'] = classifications_by_category
```

```python
def classify_according_to_threshold(threshold: int = 0.85):
  classifications_by_category = []
  for cs in classifications:
    if cs['score'] <= threshold:
      classifications_by_category.append("neutral")
    else:
      classifications_by_category.append(cs['label'].lower())

  return classifications_by_category
```

```python
df['classification 85% treshold'] = classify_according_to_threshold(0.85)
```

```python
df['classification 95% treshold'] = classify_according_to_threshold(0.95)
```

```python
df.head(10)
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, precision_recall_fscore_support

def rating_2_sentiment_label(rating: int):
  if rating == 5:
    return 'positive'
  elif 2 <= rating <= 4:
    return 'neutral'
  elif rating == 1:
    return 'negative'

df['sentiment_label'] = df['ratings'].apply(lambda x: rating_2_sentiment_label(x))
y_true = df['sentiment_label']
```

```python
print(accuracy_score(y_true, df['classification 75% treshold'].values.tolist()))
print(precision_score(y_true, df['classification 75% treshold'].values.tolist(), average=None))
print(recall_score(y_true, df['classification 75% treshold'].values.tolist(), average=None))
precision_recall_fscore_support(y_true, df['classification 75% treshold'].values.tolist(), average=None)
```

```python
print(accuracy_score(y_true, df['classification 85% treshold'].values.tolist()))
print(precision_score(y_true, df['classification 85% treshold'].values.tolist(), average=None))
print(recall_score(y_true, df['classification 85% treshold'].values.tolist(), average=None))
precision_recall_fscore_support(y_true, df['classification 85% treshold'].values.tolist(), average=None)
```

```python
print(accuracy_score(y_true, df['classification 95% treshold'].values.tolist()))
print(precision_score(y_true, df['classification 95% treshold'].values.tolist(), average=None))
print(recall_score(y_true, df['classification 95% treshold'].values.tolist(), average=None))
precision_recall_fscore_support(y_true, df['classification 95% treshold'].values.tolist(), average=None)
```

```python
# we can see that as much as we increase the treshold the results gets better, especially in the neutral category. 
# It's important to ask our selves whether we need to increase it to 99%
```

```python
df['classification 99% treshold'] = classify_according_to_threshold(0.99)
```

```python
print(accuracy_score(y_true, df['classification 99% treshold'].values.tolist()))
print(precision_score(y_true, df['classification 99% treshold'].values.tolist(), average=None))
print(recall_score(y_true, df['classification 99% treshold'].values.tolist(), average=None))
precision_recall_fscore_support(y_true, df['classification 99% treshold'].values.tolist(), average=None)
```
